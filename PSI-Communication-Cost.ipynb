For the original 1-out-of-2 OT cost, say we have $m$ pairs of $l$-bit strings, 

1. the sender to receiver communication cost is $2mk$, where $k$ is the secure parameter for OT extension. This can be further reduced to $mk$ by a PRG.
2. the next part is the receiver to sender, which is $2ml$, every time it will send a pair of $l$-bit strings. This can be further eliminated if using R-OT.

So in the optimal case, the communication cost of a single OT is $k$. For 1-out-of-N OT, the communication cost is $2k$ per OT (result from a paper I did't read yet. "imprvoted ot extensiont for transfering short secret").






## 2 Faster Private Set Intersection based on OT Extension (Usenix Security'14)
This scheme is OT-based. The basic flow is 1) a protocol for equality test 2) check if an element is in a set, 3) get set intersection of two sets.

### 2.1 The basic PEQT protocol

two $\sigma$-bit elements $x$ ($P_1$) and $y$ ($P_2$) are engaged in a random 1-out-of-2 OT.

1. For every bit, $P_1$ gets two random $l$-bit strings $\{s_0^i, s_1^i\}$, while $P_2$ get one of it based on its own bit as choice bit in the OT.
2. P1 computes $m_{p_1} = \bigoplus_{i=1}^{\sigma} s^i_{x[i]}$ and sends it to $P_2$. P2 computes its own version $m_{p_2} = \bigoplus_{i=1}^{\sigma} s^i_{y[i]}$ and decide $x=y$ if $m_{p_1} = m_{p_2} $.

The above protocol is base-2 OT. We can increase the base of OT for performance. Say we have $N = 2^{\eta}$, which is 1-out-of-N OT. $P_1$ and $P_2$ 's $\sigma$-bit elements can be represented by $t=\frac{\sigma}{\eta}$ letters of an alphabet of sizze $N$. For example, base 2, 1101, each bit could be 0 or 1. Or base 10, 13, each digit could be 0-9, which needs a 1-out-of-10 OT every time.

if $x \neq y$, then the probability of $m_{p_1} = m_{p_2}$ is $2^{-l}$. The is the same as secrect sharing. 
as long as the last string received by $P_2$, $s^t_{y[t]}$ is equal to $\bigoplus_{i=1}^{t} s^i_{x[i]} \bigoplus_{i=1}^{t-1} s^i_{y[i]} $. This event happens with probability as $2^{-l}$. So $l$ can be set as the statistical security parameter $\lambda$

**communication cost** 
Before optimization, we use 1-out-of-2 OT, communication cost is $\sigma k + 1*\lambda$, the string length is set as $l=\lambda$, which is same as the statistical security parameter. (see why $l=\lambda$ in paper.)

After optimization, there are $\frac{\sigma}{\eta}$ 1-out-of-$N$ OTs, the communication cost becomes $\frac{2 k \sigma}{\eta} + \lambda $ since each OT here will sends $2k$.


### 2.2 Private set inclusion protocol

now the new problem is given a $y$ from $P_2$ and $P_1$ has $X=\{x_1, x_2, \cdots, x_{n_1}\}$. The 1-out-of-$N$ OT is run $t$ times, string length is $n_1 l$, i.e. in each OT, $P_1$ obtains $N$ random strings $\{s_0^i,s_1^i, \cdots, s_{N-1}^i, \} \in \{0,1\}^{n_1 l}$ and $P_2$ gets the $s^i_{y[i]}$. Each string is $n_1$ substring of $l$. The following will be using step 2 as in PEQT protocol. 

If we want  $m_{p_2}$ is distinc with $m_{p_1}[j]$ for $\forall j$. this event happens with probability now as $n_1 2^{-l}$.
In order to maintain correctness with probability as $1-2^{-\lambda}$, $l$, the bit length of OT, should be set as $\lambda + \log_2 n_1 $.

Position is not disclosed when permutation is done. 

**communication cost**  
Since $l$ is changed, the communication cost becomes 
$$ \frac{2 k \sigma}{\eta} + n_1 * (\lambda + \log_2 n_1)   $$


### 2.3 The OT-based PSI protocol

Now $P_1$ has $X=\{x_1, x_2, \cdots, x_{n_1}\}$ and $P_2$ has $Y=\{y_1, y_2, \cdots, y_{n_1}\}$.

**communication cost**
$$ \frac{2 k n_2 \sigma}{\eta} + n_1 * n_2 * (\lambda + \log_2 n_1)   $$


## 3 Our scheme, a hash-based PSI protocol


Now $A$ has $S_A = \{a_1, a_2, \cdots, a_{n_a}\}$ and $B$ has $S_B = \{b_1, b_2, \cdots, b_{n_b}\}$.

### 3.1 Protocol 1
$S_D$ is in $\{0,1, \cdots, 2^k -1 \}$, $S_A, S_B \subseteq S_D $. 
Crypto hash functions:  $H: k \rightarrow k^{'}$ and $H^{'}: k^{'} \rightarrow k^{''}$.

$$ \overline{a_i} = H(a_i) \oplus r_{a:i} $$

A sends all $\overline{a_i} $ to $b$.  

**communication cost** is $n_a * k^{'}$.

Similarly,

$$ \overline{b_i} = H(b_i) \oplus r_{b:i} $$

For $(i,j)$ where $i \in[1,n_a], j \in [1,n_b]$

$$ \overline{\overline{a_{i,j}}} = H^{'} (\overline{a_i} \oplus r_{b:j})$$

B sends all $\overline{b_i}$ and $\overline{\overline{a_{i,j}}}$ to A. 

**communication cost** is $n_b * k^{'} + n_a*n_b*k^{''} $.

A: for every $(i,j)$, computes
$$ \overline{\overline{b_{i,j}}} = H^{'} (\overline{b_j} \oplus r_{a:i})$$
And check for set intersection on $ \overline{\overline{a_{i,j}}} = \overline{\overline{b_{i,j}}} $
 

(Shanhe: H need to ensure for all $b_i$ and $a_i$, there is not confliction. there is a relationship between $k^{'}$ and $k^{''}$ with $n_a$ and $n_b$ to ensure the confliction probabilities is very small. )


### 3.2 Protocol 2

Crypto hash $H^{''}$ map $S_D$ to $\{0,1, \cdots, 2^{k^{'}}-1 \}$.

a pre-defined parameter $n_g$, $n_g$ divides $2^{k^{'}}$.

For $i = 0, 1, \cdots, \frac{2^{k}}{n_g}-1 $, 

A: computes $S_{A:i} =  S_A \cap \{a | H^{''}(a) \in \{ i n_g, i n_g + 1, \cdots, (i+1) n_g -1   \}\}$.
If $|S_{A:i}| > u$, protocol fails.
O.w., add $ u -  |S_{A:i}|$ random elements whose $H^{''}$ values are out of $\{ i n_g, i n_g + 1, \cdots, (i+1) n_g -1   \}\}$ to $S_{A:i}$.

B does the same: computes $S_{B:i} =  S_B \cap \{b | H^{''}(b) \in \{ i n_g, i n_g + 1, \cdots, (i+1) n_g -1   \}\}$.
If $|S_{B:i}| > u$, protocol fails.
O.w., add $ u -  |S_{B:i}|$ random elements whose $H^{''}$ values are out of $\{ i n_g, i n_g + 1, \cdots, (i+1) n_g -1   \}\}$ to $S_{B:i}$.

A B run protocol 1 on $S_{A:i}$ and $S_{B:i}$.  

$u$ is detemined by probability inequalities. 

** communication cost ** need to figure out how many calls to Protocol 1. (Shanhe: I may be wrong on this.)

Normal case, say we set a $n_g$, then the communication cost is 
$$ \frac{2^{k}}{n_g} * (2 * n_g * k^{'} + n_g*n_g*k^{''})  =  2^{k+1}*k^{'} + n_g*2^{k}*k^{''} $$

If we apply protocol 1 directly , the communication cost is $$ ((n_a + n_b) * k^{'} + n_a*n_b*k^{''})  < 2^{k+1}*k^{'} + n_a * n_b * k^{''} < 2^{k+1}*k^{'}+2^{2k}*k^{''} $$

If $n_g = 1$, then each bin has at most one element. The protocol 2 need to run the loop for $2^{k}$ times. However, for each instance, it only cost $2 * k^{'} + k^{''}$ for communication. So the total communication cost is $$ 2^{k} * (2 * k^{'} + k^{''}) = 2^{k+1}*k^{'}+2^{k}*k^{''} $$.

If $n_g = 2^{k}$, the communication cost is  $$  (2 * 2^{k} * k^{'} + 2^{k}*2^{k}*k^{''}) = 2^{k+1}*k^{'}+2^{2k}*k^{''} $$
